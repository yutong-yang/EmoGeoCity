### 从古诗词网 https://www.gushiwen.cn/ 爬取作品表中作品的赏析，即网页中【赏】中的部分

定义爬虫函数

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options


def get_shangxi_from_crawler(browser, work_name, person_name):
    has_found = False # flag 用于标记诗词是否已经找到
    pages = ['1', '2']
    shangxi_text = '' # 记录返回值
    # 只找诗文中的作品
    for page in pages:
        if has_found:
            break
        browser.get('https://so.gushiwen.cn/search.aspx?type=title&page=' + page + '&value=' + work_name + '&valuej=' + work_name[0])
        sons_list = browser.find_elements(By.CLASS_NAME, 'sons')
        # sons 是检索出来的包括一首诗的div
        for sons in sons_list:
            if has_found:
                break
            cont = sons.find_element(By.CLASS_NAME, 'cont').find_elements(By.TAG_NAME, 'div')
            yizhu = cont[0] # 译 注 赏 背
            zhengwen = cont[1] # 正文
            # 找到当前诗的作品名
            cur_work_name = zhengwen.find_element(By.TAG_NAME, 'p').text
            # 找到当前诗的作者名
            source_p = zhengwen.find_elements(By.CLASS_NAME, 'source') # 部分诗词有两条来源信息，其中第二条是作者信息
            if len(source_p) == 1:
                source_p = source_p[0]
            else:
                source_p = source_p[1]
            cur_person_name = source_p.find_elements(By.TAG_NAME, 'a')[0].text
            if cur_work_name == work_name and cur_person_name == person_name:
                shangxi_img = yizhu.find_elements(By.TAG_NAME, 'img')[2]
                if shangxi_img.value_of_css_property('display') == 'block':
                    # 当前诗词有赏析
                    has_found = True
                    shangxi_img.click()
                    # shangxi_p = zhengwen.find_element(By.CLASS_NAME, 'contson').find_elements(By.TAG_NAME, 'p')
                    shangxi_p = zhengwen.find_element(By.CLASS_NAME, 'contson').find_element(By.CLASS_NAME, 'hr').find_elements(By.XPATH, './following-sibling::*')
                    for p in shangxi_p:
                        shangxi_text += p.text
    return shangxi_text
```

读取数据集

```python
import numpy as np
import pandas as pd
import json
data_dict = np.load('../datasets/datasets.npy', allow_pickle=True).item()
works_df = data_dict['作品表']
people_df = data_dict['人物表']
events_df = data_dict['事件表']
location_df = data_dict['地点表']
```

遍历作品表，爬取赏析

```python
from tqdm import tqdm
chrome_options = Options()
chrome_options.add_argument('--headless') # 关闭前台显示
browser = webdriver.Chrome(chrome_options=chrome_options)

work_shangxi_list = []
num = 0
for row_id, row in tqdm(works_df.iterrows()):
    work_name = row['作品名'].strip()
    work_id = row['作品id']
    person_name = row['名字']
    person_id = row['人物id']
    # 数据中没有诗歌全文的默认不是诗歌，忽略
    if pd.isna(row['诗歌全文']):
        continue
    shangxi_text = get_shangxi_from_crawler(browser, work_name.strip('《》'), person_name)
    if shangxi_text == '':
        continue
    num += 1
    work_shangxi_list.append([work_id, work_name, person_id, person_name, shangxi_text])
    # print('已找到{0}篇赏析'.format(num))
print('共找到{0}篇赏析'.format(num))
work_shangxi_df = pd.DataFrame(data=work_shangxi_list, columns=['作品id', '作品名', '人物id', '名字', '赏析'])
# np.save('../datasets/work_shangxi.npy', {'work_shangxi': work_shangxi_df})
# work_shangxi_df.to_csv('../bqs_only/work_shangxi.csv', index=False)
```

统计赏析中的词频

```python
import jieba
import numpy as np
import pandas as pd
import re
remove_chars = '[·’!"\#$%&\'()＃！（）*+,-./:;<=>?\@，。：?￥★、…．＞【】［］《》？“”‘’\[\\]^_`{|}~]+'

work_shangxi_df = np.load('../datasets/work_shangxi.npy', allow_pickle=True).item()['work_shangxi']
shangxi_freq_dict = {}
for row_id, row in work_shangxi_df.iterrows():
    cur_shangxi = row['赏析'].replace('\n', '')
    cur_shangxi = re.sub(remove_chars, '', cur_shangxi)
    cur_jieba_list = list(jieba.cut(cur_shangxi, cut_all=True))
    cur_jieba_list = filter(lambda x: len(x) > 1, cur_jieba_list)
    for word in cur_jieba_list:
        if word in shangxi_freq_dict.keys():
            shangxi_freq_dict[word] += 1
        else:
            shangxi_freq_dict[word] = 1
shangxi_freq_list = []
for word, frequency in shangxi_freq_dict.items():
    shangxi_freq_list.append([word, frequency])
shangxi_freq_df = pd.DataFrame(data=shangxi_freq_list, columns=['词语', '频率'])
shangxi_freq_df.sort_values(by='频率', ascending=False, inplace=True)
# np.save('../datasets/work_shangxi_freq.npy', {'work_shangxi_freq': shangxi_freq_df})
# shangxi_freq_df.to_csv('../bqs_only/work_shangxi_freq.csv', index=False)
```

找出分词后，表达情感的词

```python
from snownlp import SnowNLP
import numpy as np
import pandas as pd

shangxi_freq_df = np.load('../datasets/work_shangxi_freq.npy', allow_pickle=True).item()['work_shangxi_freq']
shangxi_freq_sentiments_list = []
for row_id, row in shangxi_freq_df.iterrows():
    word = row['词语']
    frequency = row['频率']
    sentiments = abs(SnowNLP(word).sentiments - 0.5) * 2
    shangxi_freq_sentiments_list.append([word, frequency, sentiments])
shangxi_freq_sentiments_df = pd.DataFrame(data=shangxi_freq_sentiments_list, columns=['词语', '频率', '是否为情感词'])
```
