把事件重要性得分加入到knowledge_map中

```python
import numpy as np
import pandas as pd
import json
data_dict = np.load('../datasets/datasets.npy', allow_pickle=True).item()
works_df = data_dict['作品表']
people_df = data_dict['人物表']
events_df = data_dict['事件表']
location_df = data_dict['地点表']
poems_df = data_dict['金陵历朝诗歌']
print(events_df.columns.tolist())
print(people_df.columns.tolist())
print(location_df.columns.tolist())
print(works_df.columns.tolist())
print(poems_df.columns.tolist())
```

```python
import random
event_score_df = pd.read_excel('../bqs_only/事件重要程度计算过程.xlsx', sheet_name=1)
with open('../json/en/knowledge_map_v2.json', 'r') as f:
    knowledge_map = json.load(f)
for event_id_str, event_obj in knowledge_map['event'].items():
    event_id = int(event_id_str.split('-')[1])
    selected_event_score_df = event_score_df[event_score_df['事件id'] == event_id]
    if selected_event_score_df.shape[0] != 0:
        knowledge_map['event'][event_id_str]['importance_score'] = float(format(float(selected_event_score_df.iloc[0]['score']), '.1f'))
    else:
        knowledge_map['event'][event_id_str]['importance_score'] = float(format(random.random() * 1.9 + 8, '.1f'))
with open('../json/en/knowledge_map_v2.json', 'w') as f:
    json.dump(knowledge_map, f)
```
